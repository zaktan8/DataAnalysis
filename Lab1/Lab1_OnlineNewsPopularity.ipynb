# ONLINE NEWS POPULARITY

## Description of the dataset

[Dataset Source](https://archive.ics.uci.edu/ml/datasets/online+news+popularity)

The goal is to predict the number of shares in social networks (popularity).

## Analysis of the dataset

```python
print(data.isnull().values.sum()) # number of NaN-values in the dataset
```

```python
categorical_columns = [c for c in data.columns if data[c].dtype.name == 'object']
numerical_columns   = [c for c in data.columns if data[c].dtype.name != 'object']
print(categorical_columns)
print(numerical_columns)
```

As shown above, all the features in the dataset are numerical. 

```python
sns.countplot(data['n_tokens_title']);
```

In this task we assume that an article can be defined as a popular one if a number of shares is greater than median of shares in the whole dataset, otherwise an unpopular.

```python
shares_median = data['shares'].median(axis=0)
shares_median
```

```python
data[data['shares'] > shares_median].count(axis=0)['shares']
```

```python
data[data['shares'] <= shares_median].count(axis=0)['shares']
```

```python
data['popularity'] = (data['shares'] > shares_median).astype(int)
```

```python
data['popularity'].value_counts().to_frame()
```

Select columns, which indicate a data channel of an article, fom initial dataset  and combine it to one categorial feature `data_channel`.

```python
data_channel_popularity = pd.DataFrame(
    data[[c for c in data.columns if 'data_channel_is' in c]].idxmax(axis=1), 
    columns=['data_channel'])
data_channel_popularity = data_channel_popularity['data_channel'].apply(lambda x: x.replace('data_channel_is_', '')).to_frame()
```

Take info about popularity of an article from initial dataset to show dependence between that feature and data channel of an article.

```python
data_channel_popularity['popularity'] = data['popularity']
```

```python
sns.countplot(x='data_channel', hue='popularity', data=data_channel_popularity);
```

As shown in the graph, the most populate fields are `tech` and `lifestyle`, and the least populate are `world` and `entertainment`.

Select columns, which indicate a day of a week, when an article was published, from initial dataset, and combine it to one categorial feature `weekday`.

```python
weekday_popularity = pd.DataFrame(
    data[[c for c in data.columns if 'weekday_is_' in c]].idxmax(axis=1),
    columns=['weekday'])
weekday_popularity = weekday_popularity['weekday'].apply(lambda x: x.replace('weekday_is_', '')).to_frame()
```

One more time take info about popularity of an article from initial dataset to show dependence between that feature and weekday.

```python
weekday_popularity['popularity'] = data['popularity']
```

```python
sns.countplot(x='weekday', hue='popularity', data=weekday_popularity);
```

In this case we can say that if an article is published at the weekend it has more chances to become popular.

# Preparation of the dataset to classification

Distinguish feature, which we want to predict - `popularity` (and also remove feature `shares`).

```python
x = data.drop(['popularity', 'shares'], axis = 1)
y = data['popularity']

feature_names = x.columns
```

Do normalization of numerical features.

```python
x = (x - x.mean(axis = 0)) / x.std(axis = 0)
```

```python
x.describe()
```

Split our data to two independent sets: train (80%) and test (20%).

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 1)

print(x_train.shape[0], x_test.shape[0])
```

## Classification

### $k$NN

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(x_train, y_train);
```

```python
y_train_predicted = knn.predict(x_train)
y_test_predicted = knn.predict(x_test)

err_train = np.mean(y_train != y_train_predicted)
err_test  = np.mean(y_test  != y_test_predicted)

print(err_train, err_test)
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_test_predicted))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_test_predicted))
```

KNN classifier shows approximately 60% of accuracy on that dataset.
